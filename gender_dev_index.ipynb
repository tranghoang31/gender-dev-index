# **Histogram and Box Plot Distribution**

```python
# Remove all missing values: na

df_cleaned = df.dropna()
```

```python
# Check if null values still exist

df_cleaned.isnull().sum().sum()
```

```python
import matplotlib.pyplot as plt

numeric_columns = df.select_dtypes(include='number')

# Create historgram graphs using dataframe with all null values removed

for column in numeric_columns.columns:
    plt.hist(df_cleaned[column].dropna(), bins=20)
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
# Create box plots

for column in numeric_columns.columns:
    plt.boxplot(df_cleaned[column].dropna())
    plt.title(f'Box plot of {column}')
    plt.show()
```
